```python
df1.drop(columns=['id'], inplace=True)
idCol = df2.id.to_numpy()
df2.drop(columns=['id'], inplace=True)
yTrain = df1.target
df1.drop(columns=['target'], inplace=True)
xTest = df2.to_numpy()
xTrain = df1.to_numpy()
xTrain.shape, yTrain.shape, xTest.shape
```

```python
def SMAPE(a, f):
    return 1/len(a) * np.sum(2 * np.abs(f-a) / (np.abs(a) + np.abs(f))*100)
```

```python
from lightgbm import LGBMRegressor

lgbmRegressor = LGBMRegressor()
lgbmRegressor.fit(xTrain, yTrain)
lgbmError = SMAPE(yTrain, lgbmRegressor.predict(xTrain))
lgbmError
```

```python
final = pd.DataFrame(columns=["id","target"])
final["id"] = idCol
final["target"] = lgbmRegressor.predict(xTest)
final
```
