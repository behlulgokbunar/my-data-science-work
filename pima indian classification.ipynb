```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
df=pd.read_csv('pima-indians-diabetes1.csv')
```

```python
df
```

```python
df.columns=['preg','plas','pres','skin','test','mass','pedi','age','class']
```

```python
df
```

```python
x=df.iloc[:,0:8]
```

```python
y=df.iloc[:,8]
```

```python
model=Sequential()
model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=150,batch_size=10,verbose=1)
```

```python
model.summary()
```

```python
scores=model.evaluate(x,y)
```

```python
print('%s: %.2f%%'%(model.metrics_names[1], scores[1]*100))
```

```python
history=model.fit(x,y,validation_split=0.20,epochs=200,batch_size=10,verbose=2)
```

```python
print(history.history.keys())
```

```python
history.history['accuracy']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig1=plt.figure(1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['training','testing'],loc='upper left')
```

```python
fig2=plt.figure(2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['training','testing'],loc='upper left')

plt.show()
```

```python
from keras.callbacks import ModelCheckpoint
```

```python
filepath='weights.best.hdf5'
checkpoint=ModelCheckpoint(filepath,monitor='val_accuracy',verbose=1,save_best_only=True,mode='max')
callbacks_list=[checkpoint]
```

```python
model.fit(x,y,validation_split=0.20,epochs=150,batch_size=10,callbacks=callbacks_list,verbose=0)
```
